```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diwali_df = pd.read_csv(r'C:\Users\arsha amal\Downloads\Diwali Sales Data.csv',encoding='unicode_escape')
diwali_df.head()
```

```python
diwali_df.describe()
```

```python
diwali_df.info()
```

```python
diwali_df.drop(['Status','unnamed1'],axis = 1,inplace = True)
```

```python
diwali_df.info()
```

```python
pd.isnull(diwali_df).sum().sum()
```

```python
diwali_df.dropna(inplace= True)
```

```python
diwali_df.head()
```

```python
GENDER 
```

```python
diwali_df['Amount'] = diwali_df['Amount'].astype('int')
```

```python
ax = sns.countplot(x = 'Gender', data = diwali_df, palette= ['pink','blue'])

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = diwali_df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending= True)
sales_gen
```

```python
sns.barplot(x = 'Gender' , y = 'Amount' , data = sales_gen, palette = ['green','yellow'])
```

```python
Most of the buyers are female
```

```python
ax = sns.countplot(x = 'Age Group' , hue = 'Gender',data = diwali_df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.countplot(x = 'Age Group' , data = diwali_df)
```

```python
sales_age = diwali_df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sales_age
sns.barplot(x = 'Age Group',y ='Amount',data = sales_age)
```

```python
#total numbers of orders from top 10 states

order_state = diwali_df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
order_state
```

```python
plt.figure(figsize = (15,5))
sns.barplot(x='State', y ='Orders',data = order_state,color = 'brown')
```

```python
#total no of amount from top 10 states

amnt_state = diwali_df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
amnt_state
```

```python
plt.figure(figsize = (15,5))
sns.barplot(x = 'State', y ='Amount',data = amnt_state)
```

```python
ax = sns.countplot(x = 'Marital_Status', data = diwali_df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
m_amnt = diwali_df.groupby(['Marital_Status', 'Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
m_amnt
```

```python
sns.barplot(x = 'Marital_Status',y = 'Amount', hue = 'Gender',data = m_amnt)
```

```python
most buyers are married women
```

```python
OCCUPATION
```

```python
plt.figure(figsize = (20,5))
ax = sns.countplot(x = 'Occupation', data = diwali_df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
o_amnt = diwali_df.groupby(['Occupation', 'Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
o_amnt
```

```python
plt.figure(figsize = (20,5))
sns.barplot(x = 'Occupation',y = 'Amount', hue = 'Gender',data = o_amnt)
```

```python
Most buyers are from IT , Aviation, Healthcare
```

```python
Product Category
```

```python
plt.figure(figsize = (30,5))
ax = sns.countplot(x = 'Product_Category', data = diwali_df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
p_amnt = diwali_df.groupby(['Product_Category', 'Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
plt.figure(figsize = (30,5))
sns.barplot(x = 'Product_Category',y = 'Amount', hue = 'Gender',data = p_amnt)
```

```python
Most buyed category are food, footwear and clothing
```
